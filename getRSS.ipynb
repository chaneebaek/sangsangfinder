```python
import feedparser
import requests
from bs4 import BeautifulSoup

rss_url = "https://www.hansung.ac.kr/bbs/hansung/2127/rssList.do?row=50"
feed = feedparser.parse(rss_url)


from urllib.parse import urlparse, urlencode, parse_qs, urlunparse

def clean_url(url):
    parsed = urlparse(url)
    params = parse_qs(parsed.query)
    params.pop("layout", None)  # layout 파라미터 제거
    new_query = urlencode({k: v[0] for k, v in params.items()})
    return urlunparse(parsed._replace(query=new_query))

def get_post_content(url):
    url = clean_url(url)  # 👈 추가
    headers = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36"}
    res = requests.get(url, headers=headers)
    soup = BeautifulSoup(res.text, "html.parser")

    content_div = soup.select_one(".txt")

    if not content_div:
        return "본문 못찾음"

    text = content_div.get_text(strip=True)
    images = content_div.find_all("img")

    if text:
        return text
    elif images:
        return "이미지만 있는 게시글"
    else:
        return "내용 없음"


for entry in feed.entries:
    print("제목:", entry.title)
    print("링크:", entry.link)

    content = get_post_content(entry.link)
    print("본문:", content[:200])  # 너무 길어서 200자만

    print("=" * 50)
```
